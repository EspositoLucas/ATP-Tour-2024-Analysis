```python
import pandas as pd
```

```python
df = pd.read_excel('../dataset/atp_2024_results_errors.xlsx')
pd.set_option('display.max_rows', None)
df.info()
```

```python
df.drop_duplicates(inplace=True)

df
```

```python
df.drop(columns=['Not Useful Column'], inplace=True)
df
```

```python
df['Winner'] = df['Winner'].str.strip("123._/")
df
```

```python
df['Country_Error'] = df['Country_Error'].str.replace(r'[^a-zA-Z0-9 ]', '', regex=True)
df
```

```python
df["Winner ATP Ranking (Current)"] =df["Winner ATP Ranking (Current)"].apply(lambda x: str(x))
df["Winner ATP Ranking (Current)"] = df["Winner ATP Ranking (Current)"].str.replace('nan', '')

df["Runner Up ATP Ranking (Current)"] =df["Runner Up ATP Ranking (Current)"].apply(lambda x: str(x))
df["Runner Up ATP Ranking (Current)"] = df["Runner Up ATP Ranking (Current)"].str.replace('nan', '')
```

```python
df["Winner ATP Ranking (Current)"]
```

```python
df["Runner Up ATP Ranking (Current)"]
```

```python
df[['Winner First Name', 'Winner Last Name']] = df['Winner'].str.split(' ',expand=True,n=1)
df[['Runner Up First Name', 'Runner Up Last Name']] = df['Runner Up'].str.split(' ',expand=True,n=1)
df
```

```python
df = df.reset_index(drop=True)
df
```

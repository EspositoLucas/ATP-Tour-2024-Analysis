## Average Prizes by Category



```python
# Average prizes by category
# Ensure 'Prize_Money' is numeric
df['Prize Money'] = pd.to_numeric(df['Prize Money'], errors='coerce')

# Drop rows with NaN values in 'Prize_Money'
cleaned_data = df.dropna(subset=['Prize Money'])

# Group by category and calculate mean
average_prizes = cleaned_data.groupby('Category')['Prize Money'].mean().sort_values()


# Plot the results
graph = average_prizes.plot(kind='barh', figsize=(20, 10), color='orange')
plt.title('Average Prizes by Tournament Category')
plt.xlabel('Average Prize')
plt.ylabel('Tournament Category')


# Add labels to bars
for container in graph.containers:
    graph.bar_label(container, fmt='%d', label_type='edge')

plt.show()
```

# Question 3: Match Results

## Set Patterns Analysis


```python
# Create a function to count sets based on the Result column
def count_sets(result):
    sets = len(result.split())
    return f"{sets} sets"

# Add a new column for number of sets played
df['Sets_Played'] = df['Result'].apply(count_sets)

# Create a pie chart for set distribution
plt.figure(figsize=(10, 6))
sets_dist = df['Sets_Played'].value_counts()
plt.pie(sets_dist, labels=sets_dist.index, autopct='%1.1f%%')
plt.title('Distribution of Matches by Number of Sets')
plt.show()
```

## Performance by Category 

```python
plt.figure(figsize=(12, 6))
category_sets = pd.crosstab(df['Category'], df['Sets_Played'])
graph = category_sets.plot(kind='bar', stacked=True)
plt.title('Match Length Distribution by Tournament Category')
plt.xlabel('Tournament Category')
plt.ylabel('Number of Matches')
plt.legend(title='Sets Played')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

## Ranking Difference Analysis


```python
# Calculate ranking difference
df['Ranking_Difference'] = abs(df['Winner ATP Ranking (Current)'] - df['Runner Up ATP Ranking (Current)'])

# Create a scatter plot
plt.figure(figsize=(12, 6))
sns.scatterplot(data=df, x='Ranking_Difference', y='Category', hue='Sets_Played')
plt.title('Impact of Ranking Difference on Match Outcome')
plt.xlabel('Difference in Rankings')
plt.ylabel('Tournament Category')
plt.legend(title='Sets Played')
plt.tight_layout()
plt.show()
```

# Tiebreak Analysis


```python
# Function to check if match had tiebreak
def has_tiebreak(result):
    return '7-6' in result or '6-7' in result

df['Has_Tiebreak'] = df['Result'].apply(has_tiebreak)

# Count tiebreak wins by player
tiebreak_winners = df[df['Has_Tiebreak']]['Winner'].value_counts().head(10)

plt.figure(figsize=(12, 6))
graph = tiebreak_winners.plot(kind='bar')
plt.title('Top 10 Players with Most Tiebreak Match Wins')
plt.xlabel('Player')
plt.ylabel('Number of Tiebreak Wins')
plt.xticks(rotation=45, ha='right')

# Add value labels
for container in graph.containers:
    graph.bar_label(container, fmt='%d')

plt.tight_layout()
plt.show()
```

## Average Prize Money by Ranking Ranges


```python
def get_ranking_range(ranking):
    if ranking <= 10:
        return 'Top 10'
    elif ranking <= 20:
        return 'Top 11-20'
    elif ranking <= 50:
        return 'Top 21-50'
    else:
        return 'Above 50'

df['Winner_Ranking_Range'] = df['Winner ATP Ranking (Current)'].apply(get_ranking_range)

avg_prize_by_ranking = df.groupby('Winner_Ranking_Range')['Prize Money'].mean().sort_values(ascending=False)

# Create a bar plot
plt.figure(figsize=(12, 6))
graph = avg_prize_by_ranking.plot(kind='bar', color='lightgreen')
plt.title('Average Prize Money by Winner\'s Ranking Range', pad=20)
plt.xlabel('Ranking Range')
plt.ylabel('Average Prize Money ($)')

# Format y-axis labels to show millions
plt.gca().yaxis.set_major_formatter(plt.FuncFormatter(lambda x, p: f'${x:,.0f}'))

# Add value labels on top of each bar
for container in graph.containers:
    graph.bar_label(container, fmt='${:,.0f}', padding=3)

# Rotate x-axis labels for better readability
plt.xticks(rotation=45, ha='right')

# Adjust layout to prevent label cutoff
plt.tight_layout()

plt.show()
```

# Question 5: Temporal Trends

## Rankings Analysis

## Category Dominance Analysis


```python
# Calculate dominance per category
dominance = df.groupby(['Category', 'Winner']).size().reset_index(name='Win_Count')

# Find all players with max wins in each category
def get_top_players(group):
    max_wins = group['Win_Count'].max()
    return group[group['Win_Count'] == max_wins]

category_dominance = dominance.groupby('Category').apply(get_top_players).reset_index(drop=True)

# Create visualization
plt.figure(figsize=(12, 8))

# Create a dictionary to track y-position offsets for each category
category_offsets = {}
offset_step = 0.3  # Aumentado para más separación vertical
category_spacing = 1.5  # Añadido espacio entre categorías

for idx, row in category_dominance.iterrows():
    category = row['Category']
    
    # Initialize or update offset for the category
    if category not in category_offsets:
        category_offsets[category] = 0
    else:
        category_offsets[category] += offset_step
    
    # Plot point with offset
    y_pos = category_dominance['Category'].unique().tolist().index(category) * category_spacing + category_offsets[category]
    
    plt.scatter(row['Win_Count'], y_pos, s=200, color='skyblue')
    
    # Add player name
    plt.annotate(f"{row['Winner']}\n({row['Win_Count']} wins)", 
                xy=(row['Win_Count'], y_pos),
                xytext=(50, 0),
                textcoords='offset points',
                va='center',
                fontsize=10)

# Adjust yticks positions to match the new spacing
plt.yticks([i * category_spacing for i in range(len(category_dominance['Category'].unique()))], 
          category_dominance['Category'].unique())

plt.title('Dominant Players by Tournament Category (2024)', pad=20)
plt.ylabel('Tournament Category')
plt.xlabel('Number of Wins')
plt.grid(True, linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

## Surfaces by Season

```python
# Convert Date column to datetime
df['Date'] = pd.to_datetime(df['Date'], format='mixed', dayfirst=True)

# Extract month from date
df['Month'] = df['Date'].dt.month

# Create a crosstab of months and surfaces
surface_by_month = pd.crosstab(df['Month'], df['Surface'])

# Combine Hard surfaces if there are duplicates
if 'Hard' in surface_by_month.columns and 'Hard ' in surface_by_month.columns:
    surface_by_month['Hard'] = surface_by_month['Hard'].add(surface_by_month['Hard '])
    surface_by_month = surface_by_month.drop('Hard ', axis=1)

# Remove columns (surfaces) that have all zeros
surface_by_month = surface_by_month.loc[:, (surface_by_month != 0).any(axis=0)]

# Create a stacked bar plot
plt.figure(figsize=(15, 8))
surface_by_month.plot(kind='bar', stacked=True)

# Customize the plot
plt.title('Distribution of Court Surfaces Throughout the Year', pad=20)
plt.xlabel('Month')
plt.ylabel('Number of Tournaments')

# Replace month numbers with month names
month_names = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 
               'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
plt.xticks(range(len(month_names)), month_names, rotation=45)

# Add legend
plt.legend(title='Surface Type', bbox_to_anchor=(1.05, 1), loc='upper left')

# Add value labels on the bars
for c in plt.gca().containers:
    # Solo añade la etiqueta si el valor no es cero
    labels = ['' if v.get_height() == 0 else str(int(v.get_height())) for v in c]
    plt.gca().bar_label(c, labels=labels, label_type='center')

# Añadir nota explicativa
plt.figtext(0.98, 0.02, 'Note: Missing values indicate zero tournaments\non that surface for that month', 
            ha='right', fontsize=8, style='italic')

# Adjust layout to prevent label cutoff
plt.tight_layout()

plt.show()
```

## Best Tournament by Country Analysis

```python
# Group by country and find the tournament with highest prize money
best_tournaments = df.groupby('Country').agg({
    'Tournament': lambda x: x[df.groupby('Country')['Prize Money'].transform('max') == df['Prize Money']].iloc[0],
    'Prize Money': 'max',
    'Category': lambda x: x[df.groupby('Country')['Prize Money'].transform('max') == df['Prize Money']].iloc[0]
}).sort_values('Prize Money', ascending=True)  # Ascending=True para que los más altos aparezcan arriba

# Crear el gráfico
plt.figure(figsize=(15, 10))
graph = best_tournaments['Prize Money'].plot(kind='barh')
plt.title('Best Tournaments by Country (Prize Money)', pad=20)
plt.xlabel('Prize Money ($)')
plt.ylabel('Country')

# Añadir etiquetas con el nombre del torneo
for i, v in enumerate(best_tournaments['Prize Money']):
    tournament_name = best_tournaments['Tournament'].iloc[i]
    category = best_tournaments['Category'].iloc[i]
    plt.text(v, i, f'  {tournament_name} ({category})', 
             va='center', fontsize=8)

# Formatear las etiquetas del eje x para mostrar millones
plt.gca().xaxis.set_major_formatter(plt.FuncFormatter(lambda x, p: f'${x:,.0f}'))

# Ajustar el layout
plt.tight_layout()
plt.show()
```
